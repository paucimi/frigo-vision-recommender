```python
%%writefile app.py
import streamlit as st
from PIL import Image
import time

#Configuraci√≥n de la p√°gina
st.set_page_config(
    page_title="The Fridge Survival",
    #page_icon="", --> icono de pagina
    layout="wide"
)


st.markdown(
    "<h1 style='text-align: center; color: white;'>The Fridge Survival</h1>"
    "<h4 style='text-align: center; color: white;'>Upload a picture of your refrigerator and we'll suggest recipes.</h4>",
    unsafe_allow_html=True)

# Creando las dos columnas
col1, col2 = st.columns(2, border=True)


# Columna 1: Subir imagen y mostrarla
with col1:
    # Subida de imagen
    uploaded_file = st.file_uploader("Upload an image to get started", type=["jpg", "png", "jpeg"])

    if uploaded_file is not None:
        # Mostrar la imagen cargada
        image = Image.open(uploaded_file)
        st.image(image, caption="uploaded image", use_container_width=False, width=750)


# Columna 2: Mostrar caracter√≠sticas (ingredientes detectados y recetas recomendadas)
with col2:
    if uploaded_file is not None:
        # Simulando proceso de ingredientes detectados
        with st.spinner(" detecting ingredients..."):
            time.sleep(2)
            ingredientes_detectados = ["tomate", "queso", "huevos"]

        st.success("ingredients correctly detected ")
        st.write("###  found ingredients:")
        for ingrediente in ingredientes_detectados:
            st.write(f"- {ingrediente}")

        # Simulaci√≥n de recomendaci√≥n de recetas
        with st.spinner(" looking for compatible recipes..."):
            time.sleep(2)
            recetas = [
                {"nombre": "Tortilla con queso", "score": 0.89},
                {"nombre": "Ensalada cesar", "score": 0.76},
                {"nombre": "Huevos al plato", "score": 0.64}
            ]

        st.write("##  recommended recipes")

        # Mostrar las recetas recomendadas con su compatibilidad
        for i, receta in enumerate(recetas):
          st.markdown(f"<div style='display: flex; justify-content: space-between;'>"
          f"<span style='font-size: 18px; font-weight: bold;'>{receta['nombre']}</span>"
          f"<span style='font-size: 16px;'>Compatibilidad: {receta['score']*100:.1f}%</span>"
          f"</div>", unsafe_allow_html=True)
          st.progress(receta["score"])

          if i < len(recetas) - 1:  # Si no es la √∫ltima receta
            st.write("---")
```
